```python
import openai
openai.organization = 'openaiì‚¬ì´íŠ¸ì—ì„œ ë°›ì€ organization nameì…ë ¥í•˜ê¸°'
openai.api_key = 'openaiì—ì„œ ë‹¤ìš´ë°›ì€ api_keyì…ë ¥í•˜ê¸°'

#aë³€ìˆ˜ ì„ ì–¸ í›„ ëŒ€í™”ë¥¼ listí˜•ìœ¼ë¡œ ëˆ„ì„, ê° ìš”ì†ŒëŠ” dictí˜•ìœ¼ë¡œ ì €ì¥
#role = ëˆ„êµ¬ì˜ ê¸°ë¡ì¸ì§€
#content = ë‚´ìš©
# a = ì €ì¥í•  ë©”ì‹œì§€ ë³€ìˆ˜
a = [{'role':'system', 'content':'dataedu has 3 offices.'}] 

# ì´ì „ ëŒ€í™” ë‚´ìš©ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
history_message = [
    {"role": "system", "content": "You are a helpful assistant."}]

# GPT-3 ì—”ì§„ ì„ íƒ
model_engine = "gpt-3.5-turbo"

# OpenAI APIë¥¼ í˜¸ì¶œí•˜ì—¬ ëŒ€í™”ë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜
def generate_chat(question):
    # OpenAI API í˜¸ì¶œí•˜ì—¬ ëŒ€í™” ìƒì„±
    history_message.append({"role":"user", "content":question})
    print(history_message)
    completions = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=history_message
    )
    message = completions.choices[0].message.to_dict()
    print(message)
    answer = message["content"].strip()

    # ì´ì „ ëŒ€í™” ë‚´ìš©ì— ìƒˆë¡œìš´ ë‹µë³€ ì¶”ê°€
    history_message.append(message)
    print(history_message)
    return answer

# ì‚¬ìš©ìì™€ì˜ ëŒ€í™” ë°˜ë³µ
while True:
    # ì‚¬ìš©ìì˜ ì…ë ¥ ë°›ê¸°
    question = input("User: ")

    # ëŒ€í™” ì¢…ë£Œ ì¡°ê±´ í™•ì¸
    if question.lower() in ["exit", "quit", "goodbye"]:
        print("Bot: Goodbye!")
        break

    # OpenAI API í˜¸ì¶œí•˜ì—¬ ë‹µë³€ ìƒì„±
    answer = generate_chat(question)

    # ì±—ë´‡ì˜ ë‹µë³€ ì¶œë ¥
    print("Bot:", answer)
```
